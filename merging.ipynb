```python
merged_1 = pd.merge(
    project,
    projectAddress,
    left_on='id',              # project's primary key
    right_on='projectId',      # address's foreign key
    how='outer'                 # Keep all projects
)
```

```python
merged_1.shape
```

```python
merged_1.head()
```

```python
merged_2 = pd.merge(
    merged_1,
    projectConfiguration,
    left_on='id_x',              # project's primary key
    right_on='projectId',      # config's foreign key
    how='outer'                 # Keep all projects
)
# Result: 60 rows (expanded due to multiple BHKs per project)
```

```python
print(merged_2.head())
```

```python
final_merged = merged_2.merge(
    projectconfigurationvariant,
    left_on='id',              # config's primary key
    right_on='configurationId', # variant's foreign key
    how='outer',
    suffixes=('_merged2', '_variant') # Keep all variants
)
# Result: 83 rows (one per property variant)
```

```python
final_merged.shape
```

```python
merged_df=final_merged.to_csv('data/5outerfinal_merged_realestate_data.csv', index=False)
```

# summary column for easy and better searching in rag

```python
df["projectSummary"] = df["projectSummary"].fillna("_")
```

```python
def build_context(row):
    return (
        f"Project Name: {row.get('projectName', '')} and type :{row.get('type', '')}. "
        f"Located at : {row.get('fullAddress', '')} , near {row.get('slug', '')} , having landmark :{row.get('landmark', '')} "
        f"Property Type: {row.get('Type', '')}. "
        f"Price: Rs{row.get('price', '')}. "
        f"having Area or carpet area: {row.get('carpetArea', '')} sq meter. "
        f"and also {row.get('projectSummary', '')} !. "
    )

df["context"] = df.apply(build_context, axis=1)
```
